```python
import pandas as pd

# Path where your Netflix dataset is stored in Databricks
csv_path = "/Volumes/workspace/default/netflix/netflix_titles.csv"

# Load dataset
df = pd.read_csv(csv_path)

# Verify successful load
display(df.head(5))
```

```python
def categorize_duration(duration):
    """
    Converts 'duration' into categories:
    Short (<60 min), Medium (60120 min), Long (>120 min), Series (for Seasons)
    Handles nulls and invalid values safely.
    """
    if pd.isna(duration):
        return "Unknown"
    
    duration = str(duration).strip()
    
    if "Season" in duration:
        return "Series"
    
    if "min" in duration:
        try:
            minutes = int(''.join([ch for ch in duration if ch.isdigit()]))  # extract digits safely
            if minutes < 60:
                return "Short"
            elif minutes <= 120:
                return "Medium"
            else:
                return "Long"
        except:
            return "Unknown"
    
    return "Unknown"

# Apply function
df["Content_Length_Category"] = df["duration"].apply(categorize_duration)

display(df[["title", "duration", "Content_Length_Category"]].head(10))
```

```python
# 🧩 Feature 2: Title Length Category (based on title character count)

df["title_length"] = df["title"].apply(lambda x: len(str(x)))

def categorize_title_length(length):
    if length < 10:
        return "Short"
    elif length <= 20:
        return "Medium"
    elif length <= 30:
        return "Long"
    else:
        return "Very Long"

df["Title_Length_Category"] = df["title_length"].apply(categorize_title_length)
display(df[["title", "title_length", "Title_Length_Category"]].head(10))
```

```python
def identify_license(desc):
    if pd.isna(desc): return "Unknown"
    if "Netflix" in desc or "NETFLIX" in desc:
        return "Original"
    else:
        return "Licensed"

df["Content_Type_Category"] = df["description"].apply(identify_license)
display(df[["title","description","Content_Type_Category"]].head(10))
```
